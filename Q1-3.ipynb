```python
df_deliveries['total_extras'] = df_deliveries[['wides', 'noballs', 'byes', 'legbyes', 'penalty']].sum(axis=1)

extras_per_match = df_deliveries.groupby('match_id')['total_extras'].sum().reset_index()
extras_per_match.rename(columns={'match_id': 'match_number', 'total_extras': 'extras'}, inplace=True)

df_deliveries['total_runs'] = df_deliveries['runs_off_bat'] + df_deliveries['total_extras']

runs_per_team = df_deliveries.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index()

runs_pivot = runs_per_team.pivot(index='match_id', columns='batting_team', values='total_runs').reset_index()

winner_list = []
looser_list = []
win_by_runs_list = []
win_by_wickets_list = []

for index, row in runs_pivot.iterrows():
    match_id = row['match_id']
    teams = row.drop('match_id').dropna()
    
    if len(teams) != 2:
        continue
    
    team1, team2 = teams.index
    score1, score2 = teams.values

    innings_info = df_deliveries[df_deliveries['match_id'] == match_id][['innings', 'batting_team']].drop_duplicates().sort_values('innings')
    first_batting_team = innings_info.iloc[0]['batting_team']
    
    if score1 > score2:
        winner = team1
        looser = team2
        if first_batting_team == team1:
            win_by_runs = score1 - score2
            win_by_wickets = np.nan
        else:
            win_by_runs = np.nan
            win_by_wickets = 10 - df_deliveries[
                (df_deliveries['match_id'] == match_id) &
                (df_deliveries['batting_team'] == team1) &
                (df_deliveries['wicket_type'].notnull())
            ]['wicket_type'].nunique()
    elif score2 > score1:
        winner = team2
        looser = team1
        if first_batting_team == team2:
            win_by_runs = score2 - score1
            win_by_wickets = np.nan
        else:
            win_by_runs = np.nan
            win_by_wickets = 10 - df_deliveries[
                (df_deliveries['match_id'] == match_id) &
                (df_deliveries['batting_team'] == team2) &
                (df_deliveries['wicket_type'].notnull())
            ]['wicket_type'].nunique()
    else:
        winner = 'Tie'
        looser = 'Tie'
        win_by_runs = 0
        win_by_wickets = 0
    
    winner_list.append({'match_number': match_id, 'winner': winner, 'looser': looser,
                        'win_by_runs': win_by_runs, 'win_by_wickets': win_by_wickets})

results_df = pd.DataFrame(winner_list)

df_matches = pd.merge(df_matches, extras_per_match, on='match_number', how='left')

df_matches = pd.merge(df_matches, results_df, on='match_number', how='left')

df_matches.head()
```

```python

teams = ['India', 'South Africa', 'Australia', 'New Zealand', 'Pakistan', 'Afghanistan', 'England', 'Bangladesh', 'Sri Lanka', 'Netherlands']

result = {
    'Team': [],
    'Matches': [],
    'Toss Won': [],
    'Toss Lost': [],
    'Matches Won After Toss Win': [],
    'Matches Won After Toss Loss': [],
    'Percentage Bat First': [],
    'Percentage Bowl First': []
}

for team in teams:
    team_matches = df_matches[(df_matches['team1'] == team) | (df_matches['team2'] == team)]
    total_matches = len(team_matches)
    toss_won = len(team_matches[team_matches['toss_winner'] == team])
    toss_lost = total_matches - toss_won
    won_after_toss_win = len(team_matches[(team_matches['toss_winner'] == team) & (team_matches['winner'] == team)])
    won_after_toss_loss = len(team_matches[(team_matches['toss_winner'] != team) & (team_matches['winner'] == team)])
    bat_first = len(team_matches[(team_matches['toss_winner'] == team) & (team_matches['toss_decision'] == 'bat')])
    bowl_first = len(team_matches[(team_matches['toss_winner'] == team) & (team_matches['toss_decision'] == 'field')])
    
    percentage_bat_first = (bat_first / toss_won) * 100 if toss_won > 0 else 0
    percentage_bowl_first = (bowl_first / toss_won) * 100 if toss_won > 0 else 0

    result['Team'].append(team)
    result['Matches'].append(total_matches)
    result['Toss Won'].append(toss_won)
    result['Toss Lost'].append(toss_lost)
    result['Matches Won After Toss Win'].append(won_after_toss_win)
    result['Matches Won After Toss Loss'].append(won_after_toss_loss)
    result['Percentage Bat First'].append(percentage_bat_first)
    result['Percentage Bowl First'].append(percentage_bowl_first)

df_result = pd.DataFrame(result)

df_result
```

```python
venue_toss_decision = df_matches.groupby(['venue', 'toss_decision']).size().unstack().fillna(0)

venue_toss_decision.plot(kind='bar', stacked=True, figsize=(12, 8), color=['#FF9999', '#66B2FF'])
plt.title('Toss Decisions at Each Venue in CWC')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.legend(title='Toss Decision', labels=['Bat', 'Field'])
plt.show()
```

```python
matches_won_after_toss = df_matches[df_matches['toss_winner'] == df_matches['winner']].groupby('venue').size()

total_matches_at_venue = df_matches.groupby('venue').size()

percentage_won_after_toss = (matches_won_after_toss / total_matches_at_venue) * 100

percentage_won_after_toss.plot(kind='bar', figsize=(12, 8), color='#66B2FF')
plt.title('Percentage of Matches Won at Each Venue After Winning the Toss')
plt.xlabel('Venue')
plt.ylabel('Percentage of Matches Won (%)')
plt.xticks(rotation=90)
plt.show()

print("Comment: The impact of winning the toss varies by venue. In some venues, the toss has a significant impact on the outcome of the match, while in others, the advantage is less pronounced.")
```

```python
import pandas as pd

teams = ['India', 'South Africa', 'Australia', 'New Zealand', 'Pakistan', 'Afghanistan', 
         'England', 'Bangladesh', 'Sri Lanka', 'Netherlands']

team_data = []

for team in teams:
    matches_played = df_matches[(df_matches['team1'] == team) | (df_matches['team2'] == team)]

    num_matches = matches_played.shape[0]
    num_won = matches_played[matches_played['winner'] == team].shape[0]
    num_lost = matches_played[matches_played['looser'] == team].shape[0]

    runs_scored = df_deliveries[(df_deliveries['batting_team'] == team) & 
                                (df_deliveries['match_id'].isin(matches_played['match_number']))]['runs_off_bat'].sum()
    overs_faced = len(df_deliveries[(df_deliveries['batting_team'] == team) & 
                                    (df_deliveries['match_id'].isin(matches_played['match_number']))]) / 6.0

    runs_conceded = df_deliveries[(df_deliveries['bowling_team'] == team) & 
                                  (df_deliveries['match_id'].isin(matches_played['match_number']))]['runs_off_bat'].sum()
    overs_bowled = len(df_deliveries[(df_deliveries['bowling_team'] == team) & 
                                     (df_deliveries['match_id'].isin(matches_played['match_number']))]) / 6.0

    nrr_for = runs_scored / overs_faced if overs_faced else 0
    nrr_against = runs_conceded / overs_bowled if overs_bowled else 0
    nrr = nrr_for - nrr_against

    extras = df_deliveries[(df_deliveries['bowling_team'] == team) & 
                           (df_deliveries['match_id'].isin(matches_played['match_number']))]['total_extras'].sum()
    percentage_extras = (extras / runs_conceded) * 100 if runs_conceded else 0

    print(f"Team: {team}, Extras: {extras}, Percentage Extras: {percentage_extras}%")

    team_data.append({
        'team': team,
        'matches_played': num_matches,
        'won': num_won,
        'lost': num_lost,
        'nrr': nrr,
        'nrr_for': nrr_for,
        'nrr_against': nrr_against,
        'extras': extras,
        'percentage_extras': percentage_extras
    })

results_df = pd.DataFrame(team_data)

results_df
```
